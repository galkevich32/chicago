```python
import pandas as pd

# 1. Load the CSV file (Assuming the first row contains your headers)
df = pd.read_csv('CookCountyForeclosures_March2026_April2026 - LISF_Final_with_mortgages.csv')

# 2. MAP THE COLUMNS (NO GLOBAL FILTER THIS TIME)
# Map Month (Index 1) and clean it up
df['Month'] = df.iloc[:, 1].astype(str).str.strip()

# Clean the dollar amount column (Index 12) to make it numeric
df['Amount'] = df.iloc[:, 12].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

# 3. CREATE MONTHLY SUBSETS (Using everything)
march_df = df[df['Month'] == 'March']
april_df = df[df['Month'] == 'April']

# --- CALCULATING THE 11 INSIGHTS ---

# 1 & 2: Total Counts (Now includes all 1,634 rows)
march_count = len(march_df)
april_count = len(april_df)

# 3: Biggest Case Value
biggest_case = df['Amount'].max()

# 4 & 5: Aggregate Mortgage Values (Sum ignores blanks)
march_sum = march_df['Amount'].sum()
april_sum = april_df['Amount'].sum()

# 6 & 7: Median Foreclosure Values (Median ignores blanks)
march_median = march_df['Amount'].median()
april_median = april_df['Amount'].median()

# 8: Overall top filer (Using the 'all_grantors' column)
overall_top_lender = df['all_grantors'].value_counts().idxmax()

# 9: Top filer in March and their count
march_top_lender = march_df['all_grantors'].value_counts().idxmax()
march_top_count = march_df['all_grantors'].value_counts().max()

# 10: Top filer in April and their count
april_top_lender = april_df['all_grantors'].value_counts().idxmax()
april_top_count = april_df['all_grantors'].value_counts().max()

# 11: Top 10 most active filers overall
top_10_filers = df['all_grantors'].value_counts().head(10)


# --- PRINTING THE RESULTS ---

print("## VOLUME & COUNTS (ALL CASE TYPES)")
print(f"1) New foreclosure lawsuits in March: {march_count:,}")
print(f"2) New foreclosure lawsuits in April: {april_count:,}")
print("-" * 50)

print("## RECORD BREAKERS")
print(f"3) Biggest case single dollar value: ${biggest_case:,.2f}")
print("-" * 50)

print("## AGGREGATE MORTGAGE VALUE")
print(f"4) Total value of March foreclosures: ${march_sum:,.2f}")
print(f"5) Total value of April foreclosures: ${april_sum:,.2f}")
print("-" * 50)

print("## MEDIAN VALUES")
print(f"6) Median foreclosure value in March: ${march_median:,.2f}")
print(f"7) Median foreclosure value in April: ${april_median:,.2f}")
print("-" * 50)

print("## LENDER ACTIVITY")
print(f"8) Most active filer overall: {overall_top_lender}")
print(f"9) Top filer in March: {march_top_lender} ({march_top_count} complaints filed)")
print(f"10) Top filer in April: {april_top_lender} ({april_top_count} complaints filed)")
print("-" * 50)

print("## 11) TOP 10 MOST ACTIVE FILERS OVERALL:")
for rank, (lender, count) in enumerate(top_10_filers.items(), 1):
    print(f"    {rank}. {lender}: {count} filings")
```

```python
import pandas as pd
import numpy as np

# 1. LOAD THE DATA
df = pd.read_csv('CookCountyForeclosures_March2026_April2026 - LISF_Final_with_mortgages.csv')

# 2. CLEAN AND PREPARE COLUMNS
# Extract Month (Index 1) and Amount (Index 12) based on your layout
df['Month'] = df.iloc[:, 1].astype(str).str.strip()
df['Amount'] = df.iloc[:, 12].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

# Clean Address to extract the City (Assumes format "STREET, CITY")
df['City'] = df['address'].astype(str).apply(lambda x: x.split(',')[-1].strip().upper() if ',' in x else 'UNKNOWN')

# Identify Corporate Defendants vs Individuals
corporate_keywords = 'LLC|INC|CORP|LTD|TRUST|COMPANY'
df['Is_Corporate'] = df['all_grantees'].astype(str).str.contains(corporate_keywords, case=False, na=False)

# Identify Standard Mortgages vs Other Liens (HOAs/Taxes)
df['Is_Mortgage'] = df['associated_docs'].astype(str).str.contains('MORTGAGE', case=False, na=False)

# --- CALCULATIONS ---

# Subsets
march_df = df[df['Month'] == 'March']
april_df = df[df['Month'] == 'April']
hoa_df = df[~df['Is_Mortgage']] # Non-mortgage cases

# Original 11 Metrics
march_count = len(march_df)
april_count = len(april_df)
biggest_case = df['Amount'].max()
march_sum = march_df['Amount'].sum()
april_sum = april_df['Amount'].sum()
march_median = march_df['Amount'].median()
april_median = april_df['Amount'].median()

overall_top_lender = df['all_grantors'].value_counts().idxmax()
march_top_lender = march_df['all_grantors'].value_counts().idxmax()
march_top_count = march_df['all_grantors'].value_counts().max()
april_top_lender = april_df['all_grantors'].value_counts().idxmax()
april_top_count = april_df['all_grantors'].value_counts().max()
top_10_filers = df['all_grantors'].value_counts().head(10)

# New Insight Calculations
top_5_cities = df['City'].value_counts().head(5)
prop_types = df['prop_type'].value_counts()
hoa_count = len(hoa_df)
top_hoa_filer = hoa_df['all_grantors'].value_counts().idxmax() if not hoa_df.empty else "N/A"
corp_count = df['Is_Corporate'].sum()
indiv_count = len(df) - corp_count

# Value Brackets
bins = [0, 200000, 500000, 1000000, np.inf]
labels = ['Under $200k', '$200k - $500k', '$500k - $1M', 'Over $1M']
df['Value_Bracket'] = pd.cut(df['Amount'], bins=bins, labels=labels)
brackets = df['Value_Bracket'].value_counts().reindex(labels)


# --- PRINTING THE COMPREHENSIVE REPORT ---

print("=" * 60)
print("             FORECLOSURE DATA INSIGHTS REPORT")
print("=" * 60)

print("\n## PART 1: ORIGINAL 11 METRICS (ALL CASES)")
print("-" * 40)
print(f"1) March Total Filings:   {march_count:,}")
print(f"2) April Total Filings:   {april_count:,}")
print(f"3) Biggest Single Case:   ${biggest_case:,.2f}")
print(f"4) March Total Value:     ${march_sum:,.2f}")
print(f"5) April Total Value:     ${april_sum:,.2f}")
print(f"6) March Median Value:    ${march_median:,.2f}")
print(f"7) April Median Value:    ${april_median:,.2f}")
print(f"8) Most Active Filer:     {overall_top_lender}")
print(f"9) Top Filer (March):     {march_top_lender} ({march_top_count} filings)")
print(f"10) Top Filer (April):    {april_top_lender} ({april_top_count} filings)")
print("\n11) Top 10 Most Active Filers Overall:")
for rank, (lender, count) in enumerate(top_10_filers.items(), 1):
    print(f"    {rank}. {lender}: {count} filings")

print("\n" + "=" * 60)
print("## PART 2: NEW POTENTIAL STORY ANGLES")
print("=" * 60)

print("\n## A. GEOGRAPHIC HOTSPOTS")
print("-" * 40)
print("Top 5 Municipalities/Neighborhoods by Volume:")
for city, count in top_5_cities.items():
    print(f"  - {city}: {count} properties")

print("\n## B. PROPERTY TYPE BREAKDOWN")
print("-" * 40)
print("Types of properties being foreclosed on:")
for p_type, count in prop_types.items():
    print(f"  - Type Code '{p_type}': {count} properties")

print("\n## C. THE HIDDEN HOA FORECLOSURES")
print("-" * 40)
print(f"Total Non-Mortgage Foreclosures (HOAs/Taxes): {hoa_count:,}")
print(f"Most active filer of these cases: {top_hoa_filer}")

print("\n## D. CORPORATE INVESTORS VS. INDIVIDUALS")
print("-" * 40)
print(f"Defendants registered as an LLC/Corp/Trust: {corp_count:,}")
print(f"Defendants registered as individuals:       {indiv_count:,}")

print("\n## E. FINANCIAL RISK BRACKETS")
print("-" * 40)
print("Breakdown of foreclosure values:")
for bracket, count in brackets.items():
    print(f"  - {bracket}: {count:,} properties")
print("=" * 60)
```
